# Distribution with disctret and continuos random Variable

```python
import numpy as np
import matplotlib.pyplot as plt

# Define coin probabilities
coin_probabilities = [0.7, 0.7, 0.7, 0.5, 0.5]  # 3 biased coins (70% heads), 2 fair coins (50% heads)

# Simulation parameters
num_simulations = 10000  # Number of simulations
results = []

# Run simulations
for _ in range(num_simulations):
    outcomes = [np.random.choice(['H', 'T'], p=[p, 1-p]) for p in coin_probabilities]
    num_heads = outcomes.count('H')
    results.append(num_heads)

# Calculate distribution
unique, counts = np.unique(results, return_counts=True)
probabilities = counts / num_simulations

# Visualization
plt.figure(figsize=(10, 6))
plt.bar(unique, probabilities, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Number of Heads (5 Coins, 3 Biased)', fontsize=16)
plt.xlabel('Number of Heads', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.xticks(range(0, 6), fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import itertools
import matplotlib.pyplot as plt
from collections import Counter

# Define the possible outcomes for a single coin flip
coin = ['H', 'T']  # H for heads, T for tails

# Generate all possible outcomes of flipping three coins
outcomes = list(itertools.product(coin, repeat=3))

# Count the number of heads in each outcome
head_counts = [outcome.count('H') for outcome in outcomes]

# Calculate the distribution
distribution = Counter(head_counts)

# Print the distribution
print("Distribution of number of heads:")
for num_heads, count in sorted(distribution.items()):
    print(f"{num_heads} heads: {count} occurrences")

# Visualize the distribution
plt.bar(distribution.keys(), distribution.values(), color='skyblue')
plt.xlabel('Number of Heads')
plt.ylabel('Occurrences')
plt.title('Distribution of Number of Heads in 3 Coin Flips')
plt.xticks(range(0, 4))
plt.show()
```

```python
# methode 2 for coin by using binoml distribution
import math
import matplotlib.pyplot as plt

def binomial_probability(n, p, k):
    """
    Calculate the probability of getting exactly k successes in n trials
    for a binomial distribution.
    """
    return math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

def visualize_binomial_distribution(n, p):
    """
    Visualize the binomial distribution for n trials and probability p.
    """
    # Possible outcomes: 0 to n successes
    outcomes = range(0, n + 1)
    # Calculate probabilities for all outcomes
    probabilities = [binomial_probability(n, p, k) for k in outcomes]

    # Plot
    plt.bar(outcomes, probabilities, color='blue', alpha=0.7)
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.xticks(outcomes)
    plt.show()

# Example usage
n = int(input("Enter the number of coin flips: "))  # Number of trials (coins)
p = 0.5  # Probability of heads for a fair coin
k = int(input("Enter the desired number of heads: "))  # Desired number of heads

# Calculate and display the probability for k successes
probability = binomial_probability(n, p, k)
print(f"The probability of getting exactly {k} heads in {n} flips: {probability:.5f}")

# Visualize the full distribution
visualize_binomial_distribution(n, p)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def visualize_coin_pmf(n, p=0.5):
    """
    Visualize the Probability Mass Function (PMF) for the number of heads in n coin flips.

    Parameters:
        n (int): Number of coin flips.
        p (float): Probability of getting heads (default is 0.5 for a fair coin).
    """
    # Define the random variable (number of heads)
    x = np.arange(0, n + 1)  # Possible outcomes: 0 to n heads
    pmf = binom.pmf(x, n, p)  # Compute PMF using the binomial distribution

    # Plot the PMF
    plt.figure(figsize=(10, 6))
    plt.bar(x, pmf, color="skyblue", edgecolor="black", alpha=0.7)

    # Add labels and title
    plt.title(f"PMF of Number of Heads in {n} Coin Flips (p = {p})", fontsize=16)
    plt.xlabel("Number of Heads", fontsize=14)
    plt.ylabel("Probability", fontsize=14)
    plt.xticks(x, fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis="y", alpha=0.6, linestyle="--")

    # Show plot
    plt.tight_layout()
    plt.show()

# Example usage
n = 10  # Number of coin flips
visualize_coin_pmf(n)
```

## Bernouli distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Parameters
n_dice = 4  # Number of dice (Bernoulli trials)
p = 1/6  # Probability of rolling a "1" (success)

# Simulate the Bernoulli distributions for 4 dice rolls
rolls = []
for i in range(1, n_dice + 1):
    roll = bernoulli.rvs(p, size=1)  # Simulate one Bernoulli trial
    rolls.append(roll[0])

# Sum of successes (number of "1s")
total_successes = np.sum(rolls)

print(f"Individual rolls (Bernoulli outcomes): {rolls}")
print(f"Total number of '1s' (successes): {total_successes}")

# Visualize Bernoulli PMFs for each die
x = [0, 1]  # Possible outcomes of a Bernoulli trial
pmf = bernoulli.pmf(x, p)

plt.figure(figsize=(10, 6))
for i in range(1, n_dice + 1):
    plt.bar(x, pmf, alpha=0.6, label=f'Die {i}', width=0.4, align='edge')
    plt.xticks(x, labels=["0 (Failure)", "1 (Success)"], fontsize=12)

plt.title("Bernoulli PMFs for Each Die Roll", fontsize=16)
plt.xlabel("Outcome (0 = Failure, 1 = Success)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc='upper left', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Continuos Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate data points for x in a specific interval
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # Interval: [μ-4σ, μ+4σ]
pdf = norm.pdf(x, mu, sigma)  # Compute PDF values

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label=f'Normal Distribution\n(μ={mu}, σ={sigma})', color='blue')
plt.fill_between(x, pdf, alpha=0.2, color='blue')  # Shade under the curve

# Highlight a specific interval (e.g., [-1σ, +1σ])
x_interval = np.linspace(mu - sigma, mu + sigma, 100)
pdf_interval = norm.pdf(x_interval, mu, sigma)
plt.fill_between(x_interval, pdf_interval, alpha=0.5, color='orange', label='Interval [-1σ, +1σ]')

# Add labels and legend
plt.title("Normal Distribution PDF", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.axvline(mu, color='red', linestyle='--', label='Mean (μ)')
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## Cumultative distribution function
The Cumulative Distribution Function (CDF) of a random variable represents the probability that the variable takes on a value less than or equal to a specific value. It is computed by integrating the Probability Density Function (PDF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # Interval: [μ-4σ, μ+4σ]
cdf = norm.cdf(x, mu, sigma)  # Compute CDF values

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(x, cdf, label=f'Normal Distribution CDF\n(μ={mu}, σ={sigma})', color='blue')
plt.axhline(1, color='green', linestyle='--', alpha=0.6, label='CDF = 1')
plt.axhline(0, color='red', linestyle='--', alpha=0.6, label='CDF = 0')

# Highlight a specific value (e.g., x = 1σ)
x_value = mu + sigma
cdf_value = norm.cdf(x_value, mu, sigma)
plt.scatter([x_value], [cdf_value], color='orange', label=f'P(X ≤ {x_value}) = {cdf_value:.2f}')
plt.axvline(x_value, color='orange', linestyle='--', alpha=0.6)

# Add labels and legend
plt.title("Cumulative Distribution Function (CDF)", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("Cumulative Probability", fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

### PDF into CDF for Continuous Random Variable

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # Interval: [μ-4σ, μ+4σ]
pdf = norm.pdf(x, mu, sigma)  # PDF values
cdf = norm.cdf(x, mu, sigma)  # CDF values

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot the PDF on the primary y-axis
color_pdf = 'blue'
ax1.plot(x, pdf, color=color_pdf, label='PDF (Probability Density)', lw=2)
ax1.set_xlabel('x', fontsize=14)
ax1.set_ylabel('PDF', color=color_pdf, fontsize=14)
ax1.tick_params(axis='y', labelcolor=color_pdf)
ax1.grid(alpha=0.3)

# Create a secondary y-axis for the CDF
ax2 = ax1.twinx()
color_cdf = 'green'
ax2.plot(x, cdf, color=color_cdf, linestyle='--', label='CDF (Cumulative Probability)', lw=2)
ax2.set_ylabel('CDF', color=color_cdf, fontsize=14)
ax2.tick_params(axis='y', labelcolor=color_cdf)

# Add legends and title
fig.suptitle('PDF and CDF of Normal Distribution', fontsize=16)
fig.tight_layout()

# Add a combined legend
lines_pdf, labels_pdf = ax1.get_legend_handles_labels()
lines_cdf, labels_cdf = ax2.get_legend_handles_labels()
ax1.legend(lines_pdf + lines_cdf, labels_pdf + labels_cdf, loc='upper left', fontsize=12)

plt.show()
```

### PMF into CDF For Discret Random Vraiable

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the outcomes of a 6-sided die
outcomes = np.arange(1, 7)  # Possible values: {1, 2, 3, 4, 5, 6}
probabilities = np.full_like(outcomes, 1/6, dtype=float)  # PMF: Equal probability for each outcome

# Compute the CDF
cdf = np.cumsum(probabilities)  # Cumulative probabilities

# Plot the PMF
plt.figure(figsize=(10, 6))
plt.bar(outcomes, probabilities, color='blue', alpha=0.7, label='PMF (Probability Mass Function)', width=0.6)
for i, p in zip(outcomes, probabilities):
    plt.text(i, p + 0.01, f'{p:.2f}', ha='center', fontsize=10)

# Plot the CDF
plt.step(outcomes, cdf, where='mid', color='green', label='CDF (Cumulative Distribution Function)', lw=2)
for i, c in zip(outcomes, cdf):
    plt.text(i, c - 0.05, f'{c:.2f}', ha='center', fontsize=10, color='green')

# Add labels, legend, and grid
plt.title('PMF and CDF for a Discrete Random Variable (6-Sided Die)', fontsize=16)
plt.xlabel('Outcome (x)', fontsize=14)
plt.ylabel('Probability', fontsize=14)
plt.xticks(outcomes)
plt.ylim(0, 1.2)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

### Standarization

Standardization is an important concept in the Normal Distribution (and many other statistical techniques) because it allows us to make meaningful comparisons between different datasets or variables, even if they have different scales or units of measurement.

Here's why standardization is important and what it entails:

1. Making Comparisons Across Different Datasets:
Datasets can have different units (e.g., one dataset in centimeters and another in inches). Without standardization, it's difficult to compare values directly.
By standardizing the data, we convert all values to a common scale (often a scale from 0 to 1, or centered around 0 with a spread of 1). This allows meaningful comparisons to be made between variables that have different units or magnitudes.
2. Standard Normal Distribution:
The Standard Normal Distribution is a normal distribution with a mean of 0 and a standard deviation of 1. This standardization process allows us to use standardized values (called z-scores) to compare how far a particular value is from the mean, relative to the spread (standard deviation).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mu = 50  # Mean of the distribution
sigma = 10  # Standard deviation of the distribution

# Generate data points for the normal distribution
data = np.random.normal(mu, sigma, 1000)

# Standardize the data using the z-score formula
standardized_data = (data - mu) / sigma

# Create the plot
plt.figure(figsize=(12, 6))

# Plot the original distribution (PDF)
plt.subplot(1, 2, 1)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, label='PDF (Normal Distribution)', color='blue', lw=2)
plt.hist(data, bins=30, density=True, alpha=0.5, color='blue')
plt.title('Original Normal Distribution (Mean=50, SD=10)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()

# Plot the standardized distribution (PDF)
plt.subplot(1, 2, 2)
x_standardized = np.linspace(-4, 4, 1000)  # Standardized data ranges from -4 to 4
pdf_standardized = norm.pdf(x_standardized, 0, 1)  # Standard Normal Distribution
plt.plot(x_standardized, pdf_standardized, label='Standardized PDF (Mean=0, SD=1)', color='green', lw=2)
plt.hist(standardized_data, bins=30, density=True, alpha=0.5, color='green')
plt.title('Standardized Normal Distribution (Mean=0, SD=1)', fontsize=16)
plt.xlabel('Standardized Value', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

### CDF


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mu = 50  # Mean of the distribution
sigma = 10  # Standard deviation of the distribution

# Generate data points for the normal distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Compute the PDF and CDF for the normal distribution
pdf = norm.pdf(x, mu, sigma)  # Probability Density Function
cdf = norm.cdf(x, mu, sigma)  # Cumulative Distribution Function

# Create the plot with two subplots
plt.figure(figsize=(12, 6))

# Plot the PDF in the first subplot
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF (Normal Distribution)', color='blue', lw=2)
plt.title('Probability Density Function (PDF)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()

# Plot the CDF in the second subplot
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF (Cumulative Distribution)', color='green', lw=2, linestyle='--')
plt.title('Cumulative Distribution Function (CDF)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Cumulative Probability', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

### PDF to CDF  with standarization

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters for the normal distribution
mu = 50  # Mean of the distribution
sigma = 10  # Standard deviation of the distribution

# Generate data points for the normal distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Compute the PDF and CDF for the normal distribution
pdf = norm.pdf(x, mu, sigma)  # Probability Density Function
cdf = norm.cdf(x, mu, sigma)  # Cumulative Distribution Function

# Standardize the data
x_standardized = (x - mu) / sigma  # Standardization using z-score
pdf_standardized = norm.pdf(x_standardized, 0, 1)  # Standardized PDF
cdf_standardized = norm.cdf(x_standardized, 0, 1)  # Standardized CDF

# Create the plot with two subplots
plt.figure(figsize=(12, 6))

# Plot the PDF in the first subplot
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF (Normal Distribution)', color='blue', lw=2)
plt.plot(x_standardized, pdf_standardized, label='Standardized PDF (Mean=0, SD=1)', color='red', lw=2, linestyle='--')
plt.title('Probability Density Function (PDF)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()

# Plot the CDF in the second subplot
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF (Normal Distribution)', color='green', lw=2)
plt.plot(x_standardized, cdf_standardized, label='Standardized CDF (Mean=0, SD=1)', color='purple', lw=2, linestyle='--')
plt.title('Cumulative Distribution Function (CDF)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Cumulative Probability', fontsize=14)
plt.grid(alpha=0.3)
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

## All Distribution in one Graph

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, norm, expon, beta, gamma

# Set up the x range
x = np.linspace(-4, 4, 1000)

# Plot Uniform Distribution
a, b = 0, 1  # range of uniform distribution
x_uniform = np.linspace(a, b, 1000)
plt.plot(x_uniform, uniform.pdf(x_uniform, loc=a, scale=b-a), label='Uniform')

# Plot Normal Distribution
mu, sigma = 0, 1  # mean and standard deviation
plt.plot(x, norm.pdf(x, mu, sigma), label='Normal')

# Plot Exponential Distribution
lambda_param = 1  # rate parameter
x_expon = np.linspace(0, 4, 1000)
plt.plot(x_expon, expon.pdf(x_expon, scale=1/lambda_param), label='Exponential')

# Plot Beta Distribution
alpha, beta_param = 2, 5
x_beta = np.linspace(0, 1, 1000)
plt.plot(x_beta, beta.pdf(x_beta, alpha, beta_param), label='Beta')

# Plot Gamma Distribution
alpha_gamma, beta_gamma = 2, 1
x_gamma = np.linspace(0, 10, 1000)
plt.plot(x_gamma, gamma.pdf(x_gamma, alpha_gamma, scale=1/beta_gamma), label='Gamma')

# Customize the plot
plt.title('Visualizing Common Continuous Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```
